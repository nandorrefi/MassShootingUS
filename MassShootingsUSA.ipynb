```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
```

```python
shootings = pd.read_csv('mass_shootings_us.csv', index_col='Incident ID')
```

```python
summary = shootings.groupby(['Year']).sum()
```

```python
dead = summary['# Killed'].values
injured = summary['# Injured'].values + summary['# Killed'].values
year = summary.index.values
```

```python
font = {'family': 'normal',
        'weight': 'normal',
        'size': 16}
matplotlib.rc('font', **font)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(16, 8)
fig.patch.set_color('lightgrey')
ax.set_axisbelow(True)
ax.grid(axis = 'y', alpha = 0.8, color='grey')
ax.bar(year, injured, 0.7, color = 'orangered')
ax.bar(year, dead, 0.7, color = 'darkred')
ax.legend(['Injured', 'Killed'])
ax.set_facecolor((0.9, 0.9, 0.9))
plt.xticks(np.arange(min(year), max(year)+1, 1))
plt.title('Mass Shooting Victims in the United States (2014-2021)')
plt.ylabel('Total Number of Victims')
fig.text(0.675, -0.025, 'Source: gunviolencearchive.org')
fig.text(0.073, -0.025, '3/20/2021')
plt.show()
```
